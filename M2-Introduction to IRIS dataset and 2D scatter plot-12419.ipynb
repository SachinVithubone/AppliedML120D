# Cumulative Distribution Function

```python
# CDF we can see what precentage of versicolor flowers have a petal length of less than 1.6?
# how to construct a CDF?
# plot for CDF to petal length
# how to read a CDF?

counts, bin_edges = np.histogram(df['petallength'], bins = 10,
                                density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);

# compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show()
```

# In above digram blue line is PDF and orange line is CDF

```python
# virginica

counts, bin_edges = np.histogram(iris_setosa['petalwidth'], bins = 10,
                                density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);

# compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show()
```
