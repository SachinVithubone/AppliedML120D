```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel('Numbers')            # Y= axis  2.0,2.5,........6.0
plt.xlabel('indcices')           # x= axis  0.0,0.5,.........3.0
plt.title('MyPlot')
plt.show()
```

```python
# 
import numpy as np
t = np.arange(0.,5.,0.2)
t
# 
plt.ylabel('Y')
plt.xlabel('X')
plt.plot(t, t**2, 'b--', label='^2')       # 'rs' , 'g', 'b--'
plt.plot(t, t**2.2,'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()

plt.legend()           # langed used for show the multiple line information
plt.show()
```

```python
x= [1,2,3,4]
y= [1,4,9,16]
plt.plot(x,y, linewidth=5.0) # its incres the line size
plt.show()
```

```python
# multiple lines

x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]

lines = plt.plot(x1,y1,x2,y2)

#
plt.setp(lines[0], color='g', linewidth=2.2) #     1both same

#
plt.setp(lines[1], 'color', 'r', 'linewidth', 2.0) #     2both same
plt.legend()
plt.grid()

#plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1= np.arange(0.0, 5.0, 0.1)
t2= np.arange(0.0, 5.0, 0.02)

plt.figure(1)
           # the  subplot()  commend specifies numrows, columns
           # fignum where fignum range from 1 to numrows*numcols

plt.subplot(211)
plt.title('SubFig1')
plt.grid()
plt.plot(t1, f(t1), 'b--')


plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.title('SubFig2')
plt.grid()
plt.show()
```

```python
plt.figure(1)                           # figure 1
plt.subplot(211)   # first subplot in figure 1
plt.plot([1,2,3])
plt.grid()

plt.subplot(212)                         # swcond subplot in figure 1
plt.plot([4,5,6])
plt.grid()


plt.figure(2)                           # second figure
plt.plot([4,5,6])                       # it create subplot (111) by default
plt.ylabel('Y')
plt.xlabel('x')
plt.title('figure2')
plt.grid()

plt.figure(1)                           # make changes in figure 1 in subplot
plt.subplot(211)
plt.plot([1,3,4])
plt.title('fig1')
plt.show()
```
